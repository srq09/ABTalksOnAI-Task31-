```python
# ============================================
# ABTalks Season-3 (AI ENGINEERING)
# Task Day-31
# Focus Area: LangChain Basics
# Tool/Tech: LangChain
# Problem Statement: Create Prompt → LLM → Output Chain
# ============================================

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# STEP 1: Put your API key here
api_key = "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"

# STEP 2: Initialize LLM
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7,
    api_key=api_key
)

# STEP 3: Prompt Template
prompt = ChatPromptTemplate.from_template(
    """
    You are a professional AI assistant.

    Explain the following topic in simple and professional language.

    Topic: {topic}

    Also provide:
    1. Definition
    2. Real-world example
    3. Key benefits
    """
)

# STEP 4: Output Parser
parser = StrOutputParser()

# STEP 5: Create Chain
chain = prompt | llm | parser

# STEP 6: User Input
user_topic = input("Enter a topic: ")

# STEP 7: Run Chain
result = chain.invoke({"topic": user_topic})

# STEP 8: Output
print("\n" + "=" * 60)
print("AI GENERATED RESPONSE")
print("=" * 60)
print(result)
print("=" * 60)

# BONUS
topics = ["Artificial Intelligence", "Machine Learning", "LangChain"]

print("\nREUSABLE AI PIPELINE DEMO\n")

for topic in topics:
    response = chain.invoke({"topic": topic})
    print(f"\nTOPIC: {topic}")
    print("-" * 50)
    print(response)
```
